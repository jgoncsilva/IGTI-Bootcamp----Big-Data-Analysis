```
**Import Libraries**
 
 
1. KMeans --> Cluster Algorithm
2. make_blobs --> Generate isotropic Gaussian blobs for clustering.
3. pyplot --> Plot lines and/or markers to the Axes. args is a variable length argument, allowing for multiple x, y pairs with an optional format string
4. pandas --> Show tables
```

```
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import pandas as pd
 
%matplotlib inline
```

```
**Values definition ​​specified by the exercise**
```

```

## values definiion for make_blobs
v_n_samples = 500
v_centers = 20
v_random_state = 999
 
## Definition values for KMeans
v_max_iter = 300
v_n_init = 10
v_init = 'k-means++'
```

```
**Generating random data and plotting an initial situation**
```

```
data,y = make_blobs(n_samples=v_n_samples, centers=v_centers, random_state=v_random_state)
 
plt.scatter(data[:, 0], data[:, 1], s=25)
plt.show()
```

```
**Generating a vector with quadratic distances with n clusters**
```

```
wcss = []
 
 
for i in range(1, 11):
    kmeans = KMeans(init=v_init, max_iter = v_max_iter, n_init=v_n_init, n_clusters=i)
    kmeans.fit(data)
    wcss.append([kmeans.inertia_]) 
 
plt.plot(range(1, 11), wcss, marker="X")
plt.xlabel('Numero de Clusters')
plt.ylabel('WCSS')
plt.show() #plota o gráfico 
 
pd.DataFrame(wcss,  range(1,11), columns=['wcss']) #monta a tabela
```

```
**Performing the processing (clustering) with the number of clusters obtained from the elbow method**
```

```
kmeans = KMeans(init=v_init, max_iter = v_max_iter, n_init=v_n_init, n_clusters=5)
 
result = kmeans.fit(data);
centers = kmeans.cluster_centers_
 
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids');
plt.legend()
plt.show()
```
